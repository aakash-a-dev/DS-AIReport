Heart attack analysis using Naiive Bayes Classifier

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import accuracy_score
import matplotlib.pyplot as plt
from sklearn.metrics import confusion_matrix
```

```python
# Load the data
data = pd.read_csv("heart.csv")
```

```python
# Split the data into training and testing datasets
x_train, x_test, y_train, y_test = train_test_split(data.iloc[:, :-1], data.iloc[:, -1], test_size=0.3, random_state=0)
```

```python
# Train the model using Naive Bayes Classifier
model = GaussianNB()
model.fit(x_train, y_train)
```

```python
# Predict the results for test dataset
y_pred = model.predict(x_test)
```

```python
# Evaluate the accuracy of the model
accuracy = accuracy_score(y_test, y_pred)
print("Accuracy: ", accuracy)
```

```python
# Evaluate the model performance using confusion matrix
cm = confusion_matrix(y_test, y_pred)
print("Confusion Matrix:\n", cm)
```

```python
# Plot the confusion matrix
plt.imshow(cm, cmap='Blues')
plt.colorbar()
plt.title("Confusion Matrix")
plt.xlabel("Predicted Label")
plt.ylabel("True Label")
plt.show()
```

```python
# Plot the results
plt.plot(y_test, label="Actual Values")
plt.plot(y_pred, label="Predicted Values")
plt.xlabel("Patient Index")
plt.ylabel("Heart Attack Risk")
plt.legend()
plt.show()
```
